```python
import requests
from bs4 import BeautifulSoup
import re


def extract_qa_from_url(url):
    html = requests.get(url, timeout=20).text
    soup = BeautifulSoup(html, "html.parser")

    # get visible text
    text = soup.get_text("\n")

    lines = [l.strip() for l in text.splitlines() if l.strip()]

    qa_pairs = []
    current_q = None

    q_pattern = re.compile(r"^(Q\d+\.|Q:)", re.IGNORECASE)
    a_pattern = re.compile(r"^(A\.|Ans:)", re.IGNORECASE)

    for line in lines:
        if q_pattern.match(line):
            current_q = line
        elif a_pattern.match(line) and current_q:
            qa_pairs.append(current_q + "\n" + line)
            current_q = None

    return "\n\n".join(qa_pairs)
```

```python
from openai import OpenAI
client = OpenAI()

def summarize_Q_A(url):
    qa_content = extract_qa_from_url(url)

    prompt = f"""
    Summarize the following FAQ in clean bullet points:

    {qa_content}
    """

    response = client.responses.create(
        model="gpt-5-mini",
        input=prompt
    )

    return response.output_text
```

```python
import re

import re

import re

def parse_qa(text):
    qa_pairs = []

    # Normalize dashes & spaces
    text = text.replace('—', '.').replace('-', '.')
    text = re.sub(r'[ \t]+', ' ', text)

    # Match Q1 Q1. Q1 - Q1 —
    question_pattern = r'(Q\s*\d+\s*[\.\:\-]?\s.*)'

    matches = list(re.finditer(question_pattern, text, re.IGNORECASE))

    print("Questions detected:", len(matches))

    if not matches:
        print("Parser failed — showing first 500 chars:")
        print(text[:500])
        return []

    for i in range(len(matches)):
        start = matches[i].start()
        end = matches[i+1].start() if i+1 < len(matches) else len(text)

        block = text[start:end].strip()

        # split answer safely
        split = re.split(r'\n\s*(A|Answer)\s*[\.\:\-]\s*', block, maxsplit=1, flags=re.IGNORECASE)

        if len(split) >= 3:
            question = split[0].strip()
            answer = split[2].strip()
        else:
            question = block
            answer = "Answer not detected"

        qa_pairs.append((question, answer))

    return qa_pairs
```

```python
from docx import Document
from docx.shared import Pt
import os

DOC_FILE = "AI_Practice_Questions.docx"


def save_to_word(qa_pairs):
    # If file exists → open it
    if os.path.exists(DOC_FILE):
        doc = Document(DOC_FILE)
    else:
        doc = Document()
        doc.add_heading("AI Interview Practice Notes", level=1)

    for q, a in qa_pairs:
        q_para = doc.add_paragraph()
        run = q_para.add_run(q)
        run.bold = True
        run.font.size = Pt(12)

        a_para = doc.add_paragraph(a)
        a_para.paragraph_format.left_indent = Pt(12)

        doc.add_paragraph("")  # spacing

    doc.save(DOC_FILE)
```

```python
import re

def extract_questions(text):
    """
    Captures only questions like:
    Q1. ....
    Q2 - ....
    Q3: ....
    Q4 — ....
    """

    # normalize weird dashes
    text = text.replace('—', '.').replace('-', '.')

    pattern = r'(Q\s*\d+\s*[\.\:\-]?\s*[^\n]+)'

    questions = re.findall(pattern, text, re.IGNORECASE)

    print("Questions found:", len(questions))
    return [q.strip() for q in questions]
```

```python
from docx import Document
import os

def save_questions_to_word(questions, filename="AI_Interview_Practice.docx"):

    if os.path.exists(filename):
        doc = Document(filename)
    else:
        doc = Document()
        doc.add_heading("AI Interview Practice Notes", level=1)

    for q in questions:
        doc.add_paragraph(q, style='List Number')

    doc.save(filename)
    print("Saved to", filename)
```
